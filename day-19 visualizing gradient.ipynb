# Visualizing a Gradient on a Contour Plot

# Create a 2D contour plot of a function like f(x,y)=x2+y2 and use a quiver plot to visualize the gradient vector at a few points.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the function f(x, y) = x + y
def f(x, y):
    return x**2 + y**2

# Step 2: Define its gradients (partial derivatives)
def grad_f(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return df_dx, df_dy

# Step 3: Create a grid of (x, y) values
x = np.linspace(-3, 3, 40)
y = np.linspace(-3, 3, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Step 4: Compute gradients on the grid
U, V = grad_f(X, Y)

# Step 5: Plot the contour of f(x, y)
plt.figure(figsize=(8, 6))
contours = plt.contour(X, Y, Z, levels=20, cmap="viridis")
plt.clabel(contours, inline=True, fontsize=8)

# Step 6: Add gradient vectors using quiver plot
plt.quiver(X, Y, U, V, color="red")

# Step 7: Highlight the origin (minimum point)
plt.scatter(0, 0, color='blue', s=80, label='Minimum (0,0)')

# Step 8: Labels and title
plt.title("Contour Plot of f(x, y) = x + y with Gradient Vectors")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True)
plt.show()
```
