# 2xcube+3y and find out del function

```python
def f(x, y):
    return 2 * x**3 + 3 * y

def grad_f(x, y):
    df_dx = 6 * x**2   # derivative w.r.t x
    df_dy = 3          # derivative w.r.t y
    return df_dx, df_dy

# Example point
x_val, y_val = 2, 1
gradient = grad_f(x_val, y_val)

print(f"f(x, y) = 2x³ + 3y")
print(f"At point (x={x_val}, y={y_val}):")
print(f"∂f/∂x = {gradient[0]}")
print(f"∂f/∂y = {gradient[1]}")
print(f"∇f = {gradient}")
```
