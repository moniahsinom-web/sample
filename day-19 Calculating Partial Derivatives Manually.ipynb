# Calculating Partial Derivatives Manually

# Write a Python function to compute the partial derivative of a simple multivariate function

```python
# Step 1: Define the multivariate function f(x, y)
def f(x, y):
    return 2 * x**2 + 3 * x * y + y**3

# Step 2a: Partial derivative of f with respect to x
def partial_x(x, y):
    # df/dx = 4x + 3y
    return 4 * x + 3 * y

# Step 2b: Partial derivative of f with respect to y
def partial_y(x, y):
    # df/dy = 3x + 3y^2
    return 3 * x + 3 * y**2

# Step 3: Test at a specific point (x=1, y=2)
x_val = 1
y_val = 2

# Compute partial derivatives at that point
df_dx = partial_x(x_val, y_val)
df_dy = partial_y(x_val, y_val)

# Display the results
print(f"Function: f(x, y) = 2x² + 3xy + y³")
print(f"At point (x={x_val}, y={y_val}):")
print(f"Partial derivative with respect to x (df/dx) = {df_dx}")
print(f"Partial derivative with respect to y (df/dy) = {df_dy}")
```
