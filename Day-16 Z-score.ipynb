# Write a Python function to calculate the Z-score for a given value, mean, and standard deviation.

```python
import numpy as np
```

# Step 1: Create a normal distribution

```python
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10
```

# Step 2: Define a function to calculate Z-score

```python
def z_score(x, mean, std):
    return (x - mean) / std
```

# Step 3: Calculate Z-score for a specific value

```python
value = 70
mean = np.mean(data)
std = np.std(data)

z = z_score(value, mean, std)

print(f"Mean of data: {mean:.2f}")
print(f"Standard Deviation: {std:.2f}")
print(f"Z-score for value {value}: {z:.2f}")
```
