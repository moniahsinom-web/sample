# Titanic Survival Analysis

# Use Seaborn visualizations to explore which factors influenced passenger survival on the Titanic

# Imported Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

titanic = sns.load_dataset("titanic")

titanic.head()
```

# Step:2. Survival Rate by Gender

```python
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title("Survival Count by Gender")
plt.show()
```

# step:3. Survival Rate by Passenger Class

```python
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title("Survival by Passenger Class")
plt.show()
```

# Step:4. Age Distribution of Survivors vs Non-Survivors

```python
sns.kdeplot(data=titanic, x="age", hue="survived", fill=True)
plt.title("Age Distribution by Survival")
plt.show()
```

# step:5. Heatmap of Correlations

```python
sns.heatmap(titanic[['survived','age','fare','pclass']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# step:6  Family Size Influence

```python
titanic["family_size"] = titanic["sibsp"] + titanic["parch"] + 1

sns.barplot(data=titanic, x="family_size", y="survived")
plt.title("Survival Rate vs Family Size")
plt.show()
```

# step:7. Fare Differences Between Survivors & Non-Survivors

```python
sns.boxplot(data=titanic, x='survived', y='fare')
plt.title("Fare Paid by Survival")
plt.show()
```
