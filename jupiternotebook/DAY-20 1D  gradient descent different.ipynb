# Visualizing Convergence with Different Learning Rates

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Common parameters
x_start = 8.0        # fixed starting point
epochs = 20
learning_rates = [0.01, 0.1, 1.0]

# Set up plot
plt.figure(figsize=(8, 5))
x_vals = np.linspace(-10, 10, 400)
plt.plot(x_vals, f(x_vals), 'k--', label='f(x) = xÂ²')

# Run gradient descent for each learning rate
for lr in learning_rates:
    x = x_start
    x_path = [x]
    y_path = [f(x)]

    for _ in range(epochs):
        x = x - lr * df(x)
        x_path.append(x)
        y_path.append(f(x))

    plt.plot(x_path, y_path, marker='o', label=f'LR = {lr}')

# Customize plot
plt.title('Effect of Learning Rate on Gradient Descent Convergence')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
