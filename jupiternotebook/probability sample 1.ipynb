# Coin Flip Simulation

```python
import random
```

# Number of flips

```python
num_flips = 1000
heads_count = 0
```

# Simulate coin flips

```python

for _ in range(num_flips):
    flip = random.randint(0, 1)  # 0 = tails, 1 = heads
    if flip == 1:
        heads_count += 1
```

# Calculate probability of heads

```python
prob_heads = heads_count / num_flips

print("Number of flips:", num_flips)
print("Number of heads:", heads_count)
print("Probability of getting heads:", prob_heads)
```
