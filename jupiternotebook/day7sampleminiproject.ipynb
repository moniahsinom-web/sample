# sales data analysis

The program calculates the total sales for each product, which is a fundamental metric for assessing a business's performance.

import libraray

```python
import pandas as pd
```

# Load the dataset

```python
data = {
    'Product': ['A', 'B', 'A', 'C', 'B'], 
    'Quantity': [10, 5, 12, 8, 7], 
    'Price': [20, 15, 20, 25, 15]
}
sales_df = pd.DataFrame(data)

print("Original Sales Data:")
print(sales_df)
```

# 2. Feature Engineering: Create a new column 'Total Sales'

# This is a vectorized operation, which is highly efficient in Pandas

```python
sales_df['Total Sales'] = sales_df['Quantity'] * sales_df['Price']

print("\nSales Data with 'Total Sales' column:")
print(sales_df)
```

# 3. Aggregation: Use groupby() to find total sales per product

# This groups the DataFrame by the 'Product' column and then sums the 'Total Sales'for each product.

```python
product_sales = sales_df.groupby('Product')['Total Sales'].sum()
```

# 4. Output: Print the result

```python
print("\nTotal Sales per Product:")
print(product_sales)
```
