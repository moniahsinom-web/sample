# Handling Missing Values with Simple Imputation

# 1. Load the Dataset

```python
import pandas as pd
import numpy as np

# Creating a sample dataset
data = {
    'age': [25, 30, np.nan, 45, 28, np.nan],
    'salary': [50000, 62000, 58000, np.nan, 76000, 54000],
    'city': ['Delhi', 'Mumbai', np.nan, 'Chennai', np.nan, 'Delhi']
}

df = pd.DataFrame(data)

print("Initial Data:")
print(df)
```

# 2. Identify Missing Values

```python
print("\nMissing Values Per Column:")
print(df.isnull().sum())
```

# 3. Handle Numerical Missing Values

# visualize skewness using a histogram

```python
import matplotlib.pyplot as plt

df['age'].hist()
plt.title("Age Distribution")
plt.show()

df['salary'].hist()
plt.title("Salary Distribution")
plt.show()
```

# 4. Handle Categorical Missing Values

```python
print(df['city'].value_counts())
```

# 5. Final Check â€” Confirm No Missing Values

```python
print("\nMissing Values After Imputation:")
print(df.isnull().sum())
```

# Final Cleaned Dataset

```python
print("\nCleaned Dataset:")
print(df)
```
