# Analyzing Tips Data (EDA Project)

# Explore the relationships between tip amount, total bill, and dining details using Seaborn visualizations.

# STEP:1. Load Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset("tips")
tips.head()
```

# STEP:2 Relationship: Total Bill vs Tip

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title("Total Bill vs Tip")
plt.show()
```

# STEP:3. Tip Percentage Column (Feature Engineering)

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill'] * 100
tips.head()
```

# STEP:4. Distribution of Tip Percentage

```python
sns.histplot(data=tips, x='tip_pct', kde=True)
plt.title("Distribution of Tip Percentage")
plt.show()
```

# STEP:5. Tips by Gender

```python
sns.boxplot(data=tips, x='sex', y='tip')
plt.title("Tip Amount by Gender")
plt.show()
```

# STEP:6. Tips by Day of Week

```python
sns.barplot(data=tips, x='day', y='tip')
plt.title("Average Tip by Day")
plt.show()
```

# STEP:7. Tips by Smoking Status

```python
sns.boxplot(data=tips, x='smoker', y='tip')
plt.title("Tips vs Smoker/Non-Smoker")
plt.show()
```

# STEP:8. Tips by Time (Lunch vs Dinner)

```python
sns.boxplot(data=tips, x='time', y='tip')
plt.title("Tips at Lunch vs Dinner")
plt.show()
```

# STEP:9. Tips by Group Size

```python
sns.barplot(data=tips, x='size', y='tip')
plt.title("Tips by Group Size")
plt.show()
```
