```python
import numpy as np
```

# 1. Define Inputs

```python
input_vector = np.array([2, 3, 4])   # 3 input values
```

# 2. Define Weights and Bias

```python
weights_matrix = np.random.rand(3, 2)  # (3 inputs, 2 outputs)
bias_vector = np.array([0.1, 0.2])     # bias for each output neuron
```

# 3. Perform Linear Transformation

```python

linear_output = input_vector @ weights_matrix + bias_vector
```

# 4. Define Activation Function (Sigmoid)

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Apply activation function
activated_output = sigmoid(linear_output)
```

# 5. Output Results

```python
print("Input Vector:\n", input_vector)
print("\nWeights Matrix:\n", weights_matrix)
print("\nBias Vector:\n", bias_vector)
print("\nLinear Output (Before Activation):\n", linear_output)
print("\nActivated Output (After Sigmoid):\n", activated_output)
```
