# Employee Performance Analysis

Merge a dataset of employee details with a performance review
dataset and calculate the average performance score per department

 import the pandas library

```python
import pandas as pd
```

# Step 2: Create sample datasets

```python
# Employee details dataset
employee_details = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['Sales', 'Engineering', 'Sales', 'HR', 'Engineering']
})

# Performance review dataset
performance_reviews = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'Performance_Score': [95, 88, 76, 92, 85]
})

print("Employee Details:")
print(employee_details)
print("\nPerformance Reviews:")
print(performance_reviews)
```

# Step 3: Merge the two datasets

```python
merged_df = pd.merge(employee_details, performance_reviews, on='employee_id')

print("\nMerged DataFrame:")
print(merged_df)
```

# Step 4: Calculate the average performance score per department

```python
average_performance_by_dept = merged_df.groupby('Department')['Performance_Score'].mean()

print("\nAverage Performance Score by Department:")
print(average_performance_by_dept)
```
