# Data centering and scaling

 Center and scale a dataset (matrix) to have a mean of 0 and a
standard deviation of 1 for each column

improt libraray

```python
import numpy as np
```

```python
# Create a sample 2D matrix (e.g., 5 rows, 3 columns)
# The random numbers have different means and scales
X = np.array([
    [10, 200, 3],
    [12, 180, 4],
    [8, 220, 2],
    [15, 250, 6],
    [9, 190, 3]
])

print("Original Matrix (X):")
print(X)
```

```python
# Calculate the mean of each column
column_means = X.mean(axis=0)
print("\nMean of each column:")
print(column_means)
```

```python
# Calculate the standard deviation of each column

column_stds = X.std(axis=0)

print("\nStandard deviation of each column:")

print(column_stds)
```
