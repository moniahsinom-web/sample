# Basic Subplot Visualization

Create a figure with two subplots: one for a line chart and one
for a scatter plot of the same data

Import library

```python
import matplotlib.pyplot as plt
import numpy as np
```

# sample Data for plotting

```python
# Optional: Add the magic command for some versions of Jupyter to ensure plots are displayed inline.
# %matplotlib inline

# 1. Generate some sample data for plotting
x = np.linspace(0, 10, 50)
y = np.sin(x)
```

# step:2 subplots

```python
# 2. Create a figure with two subplots arranged in 1 row and 2 columns
# The function returns the figure and an array of axes.
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
```

Step:3 line chart

```python
# 3. Plot a line chart on the first subplot (ax1)
ax1.plot(x, y, color='blue', label='sin(x) line')
ax1.set_title('Line Chart')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.grid(True)
ax1.legend()
```

```python
# 4. Plot a scatter plot on the second subplot (ax2)
ax2.scatter(x, y, color='red', label='sin(x) points')
ax2.set_title('Scatter Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.grid(True)
ax2.legend()

```

```python
# 5. Automatically adjust subplot parameters for a tight layout
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(np.random.rand(10))
ax[0, 1].plot(np.random.rand(10))
ax[1, 0].plot(np.random.rand(10))
ax[1, 1].plot(np.random.rand(10))

plt.tight_layout(h_pad=1.0, w_pad=1.5) # Adjust horizontal and vertical spacing between subplots
plt.show()
```

```python
# 6. Display the figure in the notebook
# In most modern Jupyter environments, this line is not strictly necessary.
plt.show()
```
