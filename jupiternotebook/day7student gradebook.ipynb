## Student Grade book

 Create a DataFrame for student grades and calculate the average
score for each student and each subject

# step:1 set up the imported libraray

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set display options to show all columns
pd.set_option('display.max_columns', None)

print("Libraries imported successfully.")
```

# step:2 create the Gradebook dataframe

```python
grade_data = {
    'Student_Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Quiz_1': [85, 92, 78, 88],
    'Quiz_2': [90, 85, 82, 95],
    'Exam_1': [95, 88, 85, 91],
    'Homework_1': [100, 95, 88, 92],
    'Homework_2': [98, 90, 85, 96]
}

gradebook_df = pd.DataFrame(grade_data)
print("Initial Gradebook:")
print(gradebook_df)
```

# step:3 calculate total and final scores

```python
# Calculate total quiz score by summing the relevant columns
gradebook_df['Total_Quizzes'] = gradebook_df['Quiz_1'] + gradebook_df['Quiz_2']

# Define weighting for final grade calculation
weight_quiz = 0.30
weight_homework = 0.30
weight_exam = 0.40

# Calculate a weighted final grade
gradebook_df['Final_Grade'] = (
    (gradebook_df['Quiz_1'] + gradebook_df['Quiz_2']) / 2 * weight_quiz +
    (gradebook_df['Homework_1'] + gradebook_df['Homework_2']) / 2 * weight_homework +
    gradebook_df['Exam_1'] * weight_exam
)
print("\nGradebook with calculated scores:")
print(gradebook_df)
```

# step:4 Assign Letter grades

```python
# Define a function to determine letter grade
def assign_letter_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# Apply the function to create a new 'Letter_Grade' column
gradebook_df['Letter_Grade'] = gradebook_df['Final_Grade'].apply(assign_letter_grade)
print("\nFinal Gradebook with letter grades:")
print(gradebook_df)
```

# step:5 save the Grade book to a file

```python
# Save the final gradebook to a CSV file
gradebook_df.to_csv('student_gradebook.csv', index=False)
print("\nGradebook saved to 'student_gradebook.csv'")
```

# step:6 Analyze Grade

```python
# Sort the gradebook by final grade in descending order
sorted_gradebook = gradebook_df.sort_values(by='Final_Grade', ascending=False)
print("\nGradebook sorted by Final_Grade:")
print(sorted_gradebook)

# Get descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(gradebook_df.describe())
```
