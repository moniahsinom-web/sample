# K-Means Clustering for Customer Segmentation

# Group customers into distinct segments based on their purchasing behavior (e.g., annual income and spending score).

# Imported Library

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

# Step 2: Simulate or Load Customer Data

```python
# Create synthetic data: 200 customers, 2 features (e.g., Annual Income & Spending Score)
X, _ = make_blobs(n_samples=200, centers=4, cluster_std=1.2, random_state=42)
```

# Step 3: Visualize the Data

```python
plt.scatter(X[:, 0], X[:, 1], c='gray')
plt.title("Customer Data Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

# Step 4: Apply K-Means Clustering

```python
# Create K-Means model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)

# Fit the model to data
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

# step-5 Visualize the Clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X')
plt.title("Customer Segments (K-Means Clustering)")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

# step-6 Determine Optimal K (Using Elbow Method)

```python
inertia = []  # sum of squared distances for each k
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(k_values, inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()
```
