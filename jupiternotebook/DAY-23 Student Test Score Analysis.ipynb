# Student Test Score Analysis

#### Explore a student test score dataset to find relationships between math, reading, writing scores and demographic factors such as gender, parental education, lunch type, and test preparation course

# 1. Load the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("student_scores.csv")
df.head()
```

# 2.Exploratory Data Analysis (EDA)

# A. Distribution of Scores

```python
sns.histplot(df["math_score"], kde=True)
plt.title("Math Score Distribution")
plt.show()
```

```python
sns.histplot(df["reading_score"], kde=True)
plt.title("Reading Score Distribution")
plt.show()
```

# B. Gender vs Scores

```python
sns.barplot(data=df, x="gender", y="math_score")
plt.title("Math Score by Gender")
plt.show()
```

# c. Test Preparation Influence

```python
sns.boxplot(data=df, x="test_prep", y="math_score")
plt.title("Math Scores vs Test Preparation")
plt.show()
```

# D. Correlation Heatmap

```python
sns.heatmap(df[["math_score","reading_score","writing_score"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Scores")
plt.show()
```
