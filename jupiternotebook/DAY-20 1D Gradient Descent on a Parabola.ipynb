# 1D Gradient Descent on a Parabola

# EXPLAINATION

```python
import numpy as np
import matplotlib.pyplot as plt

# 1️⃣ Define the function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# 2️⃣ Initialization
x_start = np.random.uniform(-10, 10)   # random start point
x = x_start

# 3️⃣ Parameters
learning_rate = 0.1
epochs = 20

# For visualization
x_history = [x]
y_history = [f(x)]

# 4️⃣ Gradient Descent Loop
for i in range(epochs):
    grad = df(x)
    x = x - learning_rate * grad   # update rule
    x_history.append(x)
    y_history.append(f(x))

# 5️⃣ Visualization
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label='f(x) = x²', color='blue')
plt.scatter(x_history, y_history, color='red')
plt.plot(x_history, y_history, color='red', linestyle='--', label='Descent Path')
plt.title("1D Gradient Descent on a Parabola")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Starting point: {x_start:.4f}")
print(f"Final point after {epochs} epochs: {x:.4f}")
```
