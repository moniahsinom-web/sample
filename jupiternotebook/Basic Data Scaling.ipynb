```python
import numpy as np
```

# 1. Create a sample 2D dataset matrix (rows = samples, columns = features)

```python
M = np.array([
    [65, 170, 75],
    [70, 180, 80],
    [60, 165, 70],
    [75, 175, 85]
])

print("Original Matrix:\n", M)
```

# 2. Compute mean and standard deviation for each feature (column-wise)

```python
mean = M.mean(axis=0)
std = M.std(axis=0)

print("\nMean of each feature:\n", mean)
print("Standard deviation of each feature:\n", std)
```

# 3. Perform scaling: (M - mean) / std

```python
M_scaled = (M - mean) / std
```

# 4. Output the scaled matrix

```python
print("\nScaled Matrix (Standardized Data):\n", M_scaled)
```
