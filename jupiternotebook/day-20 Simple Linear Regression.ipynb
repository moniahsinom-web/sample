# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# 1 Generate some sample data (y = 4x + 2 + noise)
np.random.seed(10)
X = np.random.rand(100, 1) * 10
y = 4 * X + 2 + np.random.randn(100, 1) * 2

# 2 Initialize parameters
m = 0.0   # slope
c = 0.0   # intercept
learning_rate = 0.001
epochs = 1000

# For plotting the loss
loss_history = []

# 3 Gradient Descent Loop
N = len(X)
for epoch in range(epochs):
    y_pred = m * X + c
    error = y - y_pred
    
    # Compute gradients
    dm = (-2/N) * np.sum(X * error)
    dc = (-2/N) * np.sum(error)
    
    # Update parameters
    m = m - learning_rate * dm
    c = c - learning_rate * dc
    
    # Compute and record loss
    loss = np.mean(error**2)
    loss_history.append(loss)
    
    # Print progress every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss={loss:.4f}, m={m:.3f}, c={c:.3f}")

# 4 Plot regression line vs data
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, m * X + c, color='red', label=f'Fitted line: y={m:.2f}x+{c:.2f}')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression using Gradient Descent')
plt.legend()
plt.show()

# 5 Plot loss curve
plt.figure(figsize=(6,4))
plt.plot(loss_history)
plt.title("Loss Function Convergence")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()

print(f"\nFinal parameters:\nm = {m:.3f}, c = {c:.3f}")
```
