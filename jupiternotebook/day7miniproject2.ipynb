# Customer Purchase History

Create a DataFrame of customer purchases and find the total
spending for each customer and the most popular product.

import libraray

```python
import pandas as pd
```

# Sample data for customer purchases

```python
data = {
    'CustomerID': [1, 2, 1, 3, 2, 1, 3],
    'Product': ['Laptop', 'Mouse', 'Mouse', 'Keyboard', 'Laptop', 'Laptop', 'Mouse'],
    'Amount': [1200, 25, 25, 75, 1200, 1200, 25]
}
```

# Create the DataFrame

```python
df = pd.DataFrame(data)
print("Original Customer Purchase History:")
print(df)
```

# Calculate total spending per customer

```python
total_spending = df.groupby('CustomerID')['Amount'].sum()

print("\nTotal spending for each customer:")

print(total_spending)
```

# Find the count of each product

```python
product_counts = df['Product'].value_counts()

print("\nPopularity of each product:")

print(product_counts)
```

# Get the most popular product


```python
most_popular_product = df['Product'].value_counts().idxmax()
print("\nThe most popular product is:", most_popular_product)
```
