# Sales Data Analysis with Bar and Pie Charts

Visualize total sales per product category using a bar chart and the
market share of each category using a pie chart.

Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# DataFrame

```python

# 1. Create DataFrame
data = {'Category': ['Electronics', 'Clothing', 'Home Goods', 'Books', 
                     'Electronics', 'Clothing', 'Electronics'],
        'Sales': [1500, 800, 1200, 500, 2000, 1000, 1800]}

df = pd.DataFrame(data)

# 2. Aggregate Data
aggregated_sales = df.groupby('Category')['Sales'].sum()
```

# subplots

```python
# 5. Subplots: Create a figure with two subplots (one row, two columns)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
```

# Create bar chart

```python
# 3. Bar Chart: Use the first subplot (axes[1])
axes[0].bar(aggregated_sales.index, aggregated_sales.values, color='yellow')
axes[0].set_title('Total Sales per Product Category')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Total Sales')
axes[0].tick_params(axis='x', rotation=45)
```

# Create pie chart

```python
# 4. Pie Chart: Use the second subplot (axes[1])
axes[1].pie(aggregated_sales.values, labels=aggregated_sales.index, autopct='%1.1f%%', startangle=90, 
            colors=["#121111","#c30505e0","#9e99ff","#99ff9b"])
axes[1].set_title('Market Share of Each Category')
```

# Display chart

```python
# Display the chart
# Reduce size of layout and show the plot
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(np.random.rand(10))
ax[0, 1].plot(np.random.rand(10))
ax[1, 0].plot(np.random.rand(10))
ax[1, 1].plot(np.random.rand(10))

plt.tight_layout(pad=1.0) # Decreasing padding around the figure
plt.show()
```
