# Basic EDA on a Clean Dataset (Iris Dataset)

# Perform a basic exploratory data analysis on the well-known Iris dataset to understand its structure, distributions, and relationships between variables.

# Step 1: Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# Step 2: Load the Dataset

```python
# Load the iris dataset from scikit-learn
iris_data = load_iris()

# Convert to a pandas DataFrame
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add the target (species) column
iris['species'] = iris_data.target

# Map numeric target to actual species names
iris['species'] = iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

# Step 3: Initial Inspection

```python
# First 5 rows
print("First 5 rows of dataset:")
print(iris.head())

# Last 5 rows
print("\nLast 5 rows of dataset:")
print(iris.tail())

# Info about data types and missing values
print("\nDataset Info:")
print(iris.info())

# Statistical summary
print("\nStatistical Summary:")
print(iris.describe())
```

# Step 4: Explore Categorical Features

```python
# Count of each flower species
print("\nValue Counts for each species:")
print(iris['species'].value_counts())
```

# Step 5: Visualize Relationships

```python
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle("Pair Plot of Iris Features", y=1.02)
plt.show()
```

```python
# Melt the dataset for easier plotting
iris_melted = iris.melt(id_vars='species', var_name='Feature', value_name='Value')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Feature', y='Value', hue='species', data=iris_melted)
plt.title("Box Plot of Iris Features by Species")
plt.xticks(rotation=45)
plt.show()
```
