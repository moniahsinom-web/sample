# sales Data Analysis

# Imported Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the Dataset

```python
# Sample sales data
data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Laptop', 'Mobile', 'TV', 'TV', 'Mobile', 'Tablet', 'Laptop'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'East'],
    'Units_Sold': [120, 340, 230, 150, 270, 190, 220, 310, 260, 180],
    'Unit_Price': [800, 400, 300, 800, 400, 500, 500, 400, 300, 800]
}

sales = pd.DataFrame(data)

# Compute revenue
sales['Revenue'] = sales['Units_Sold'] * sales['Unit_Price']
```

# Step 3: Inspect the Data

```python
print("First few rows:")
print(sales.head())

print("\nDataset Info:")
print(sales.info())

print("\nStatistical Summary:")
print(sales.describe())
```

```python
# Total units sold per product
top_products_units = sales.groupby('Product')['Units_Sold'].sum().sort_values(ascending=False)

# Total revenue per product
top_products_revenue = sales.groupby('Product')['Revenue'].sum().sort_values(ascending=False)

print("\nTop-Selling Products (by Units):")
print(top_products_units)

print("\nTop-Selling Products (by Revenue):")
print(top_products_revenue)
```

# Step 5: Visualize Top-Selling Products
# (a) By Units Sold

```python
plt.figure(figsize=(6,4))
sns.barplot(x=top_products_units.index, y=top_products_units.values, palette='viridis')
plt.title("Top-Selling Products by Units Sold")
plt.ylabel("Units Sold")
plt.xlabel("Product")
plt.show()
```

# (b) By Revenue

```python
plt.figure(figsize=(6,4))
sns.barplot(x=top_products_revenue.index, y=top_products_revenue.values, palette='magma')
plt.title("Top-Selling Products by Total Revenue")
plt.ylabel("Total Revenue")
plt.xlabel("Product")
plt.show()
```

# Step 6: Find Average Revenue per Region

```python
avg_revenue_region = sales.groupby('Region')['Revenue'].mean().sort_values(ascending=False)
print("\nAverage Revenue per Region:")
print(avg_revenue_region)
```

# Step 7: Visualize Average Revenue per Region

```python
plt.figure(figsize=(6,4))
sns.barplot(x=avg_revenue_region.index, y=avg_revenue_region.values, palette='coolwarm')
plt.title("Average Revenue per Region")
plt.ylabel("Average Revenue")
plt.xlabel("Region")
plt.show()
```
