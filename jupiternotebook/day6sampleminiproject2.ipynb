# Step 2: Perform SVD

```python
("arr") U , S , Vt == np.linalg.svd ("gray_image", full_matrices = False)
```

# Step 3: Function to reconstruct image using top k singular values

```python
def reconstruct_image(k):
    S_k = np.diag(S[:k])
    return U[:, :k] @ S_k @ Vt[:k, :]
```

# Step 4: Display results for different k values

```python
k_values = [10, 50, 100, 200]
plt.figure(figsize=(10, 6))
```

```python
for i, k in enumerate(k_values):
    def reconstruct_image(k):
       plt.subplot (2, 2, i+1)
       plt.imshow (cmap='gray')
       plt.title(f'k = {k}')
       plt.axis('off')

       plt.suptitle('Image Compression using SVD', fontsize=14)
       plt.tight_layout()
       plt.show()
```
