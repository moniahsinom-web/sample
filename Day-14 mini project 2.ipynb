# Deck of Cards Probability


#  Calculate the probability of drawing a red card given that the card is a face card

```python
def conditional_probability(p_a_and_b, p_b):
    """
    Calculate conditional probability P(A|B) = P(A and B) / P(B)
    """
    if p_b == 0:
        return "Error: Probability of event B cannot be zero."
    return p_a_and_b / p_b
```

```python
# Example: Deck of 52 cards
# There are 12 face cards in total (Jack, Queen, King of each suit)
# There are 6 red face cards (Jack, Queen, King of Hearts and Diamonds)

red_face_cards = 6        # P(Red AND Face)
total_face_cards = 12     # P(Face)
```

# Since these are counts, we can use them directly

```python
p_a_and_b = red_face_cards / 52   # Probability of drawing a red face card
p_b = total_face_cards / 52       # Probability of drawing a face card

result = conditional_probability(p_a_and_b, p_b)
print(f"P(Red | Face) = {result}")
```
