# f(x,y)=4x2y+2y3x

```python
import sympy as sp

# Define the variables
x, y = sp.symbols('x y')

# Define the function
f = 4*x**2*y + 2*y**3*x

# Compute partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

# Display the results
print("∂f/∂x =", df_dx)
print("∂f/∂y =", df_dy)

# Gradient vector
grad_f = (df_dx, df_dy)
print("∇f =", grad_f)
```
