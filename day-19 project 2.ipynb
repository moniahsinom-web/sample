# Simple Gradient Descent for Linear Regression

#  Implement a basic linear regression model from scratch using gradient descent.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 100)
y = 3 * X + 4 + np.random.randn(100) * 2   # true slope=3, intercept=4, with noise

# Step 2: Initialize parameters
m = 0.0  # slope
b = 0.0  # intercept
alpha = 0.01  # learning rate
epochs = 1000  # number of iterations

# Step 3: Gradient Descent loop
N = len(X)
for i in range(epochs):
    y_pred = m * X + b
    error = y - y_pred
    dm = -(2/N) * np.dot(X, error)
    db = -(2/N) * np.sum(error)
    
    # Update parameters
    m -= alpha * dm
    b -= alpha * db
    
    # Optional: Print progress every 100 iterations
    if i % 100 == 0:
        cost = np.mean(error ** 2)
        print(f"Epoch {i}: m={m:.3f}, b={b:.3f}, Cost={cost:.3f}")

# Step 4: Plot results
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Data", color='blue')
plt.plot(X, m * X + b, color='red', label=f"Fitted line: y={m:.2f}x+{b:.2f}")
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(f"\nFinal parameters: m={m:.2f}, b={b:.2f}")
```
