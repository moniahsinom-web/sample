```python
import numpy as np
import matplotlib.pyplot as plt
```

# Step 1: Generate synthetic data

```python
np.random.seed(42)
X = np.random.rand(100, 1)  # 100 samples, 1 feature
y = 3 * X.flatten() + 2 + np.random.randn(100) * 0.3  # true relation + noise
```

# Step 2: Add bias term (column of 1s)

```python
X_b = np.c_[np.ones((100, 1)), X]  # shape (100, 2)
```

# Step 3: Compute parameters using Normal Equation

```python
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print("Optimal Parameters (θ):", theta_best)
```

# Step 4: Predict values

```python
X_new = np.array([[0], [1]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred = X_new_b @ theta_best
```

# Step 5: Visualization

```python
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Best Fit Line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Normal Equation")
plt.legend()
plt.show()
```
