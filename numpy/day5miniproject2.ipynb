```python
import numpy as np
```

# Step 1: Create a sample 2D grayscale "image" array

```python
image = np.array([
    [10, 20, 30, 40],
    [20, 30, 40, 50],
    [30, 40, 50, 60],
    [40, 50, 60, 70]
], dtype=float)

print("Original Image:\n", image)
```

# Step 2: Prepare an empty array for the blurred result

```python
blurred = np.zeros_like(image)
```

# Step 3: Apply blur using a 3x3 average filter (manual convolution)

```python
for i in range(1, image.shape[0] - 1):        # avoid borders
    for j in range(1, image.shape[1] - 1):
        # extract 3x3 neighborhood
        neighborhood = image[i-1:i+2, j-1:j+2]
        # compute average
        blurred[i, j] = np.mean(neighborhood)
```

# Step 4: Display the blurred image

```python
print("\nBlurred Image:\n", blurred)
```
