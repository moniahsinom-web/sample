```python
import numpy as np
```

# Step 1: Create a sample 2D array

```python
arr = np.array([[10, 20, 30],
                [15, 25, 35],
                [20, 30, 40],
                [25, 35, 45]])

print("Original Array:\n", arr)
```

# Step 2: Compute mean and std for each column (axis=0)

```python
mean = arr.mean(axis=0)
std = arr.std(axis=0)

print("\nColumn-wise Mean:\n", mean)
print("\nColumn-wise Standard Deviation:\n", std)
```

# Step 3: Normalize the data

```python
normalized_arr = (arr - mean) / std
```

# Step 4: Display normalized data

```python
print("\nNormalized Array (Z-score normalization):\n", normalized_arr)
```
