```python
import numpy as np
```

```python
# 1. Create a Sample Array
# Generate a 2D array with shape (5, 4) and random integers between 10 and 100
arr = np.random.randint(10, 100, size=(5, 4))
print("Original Array:\n", arr)
```

# 2. Calculations
# Mean of each column

```python

col_mean = arr.mean(axis=0)
print("\nMean of each column:", 
col_mean)
```

# Maximum value in each row

```python
row_max = arr.max(axis=1)
print("Maximum value in each row:", row_max)
```

# Total sum of all elements

```python
total_sum = arr.sum()
print("Total sum of all elements:",
 total_sum)
```

# 3. Indexing
# Select first two rows and last two columns

```python
sliced_arr = arr[0:2, 2:4]
print("\nFirst two rows and last two columns:\n",
 sliced_arr)
```

# Boolean indexing â€” elements greater than 50

```python
greater_than_50 = arr[arr > 50]
print("Elements greater than 50:", greater_than_50)
```

# 4. Output

```python
print("\nSummary of Calculations:")
print("------------------------")
print(f"Column Means: {col_mean}")
print(f"Row Maximums: {row_max}")
print(f"Total Sum: {total_sum}")
print(f"Sliced Array (0:2, 2:4):\n{sliced_arr}")
print(f"Elements > 50: {greater_than_50}")
```
