```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# This command is typically needed in older versions of Jupyter
# to ensure plots are displayed inline.
%matplotlib inline
```

# Step 1: Load the data

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

# step:2 Survival bar plot

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.show()
```

# Step 3: Survival by age and gender (Violin plot)

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='sex', y='age', hue='survived', data=titanic, split=True)
plt.title('Age Distribution and Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
