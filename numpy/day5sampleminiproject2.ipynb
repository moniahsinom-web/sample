# Simple Image Manipulation

To load a grayscale image (simulated as a NumPy array) and perform simple transformations like inversion and cropping. 

Import Libraray

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load Fake Image

```python

# 1. Load an Image (simulated as a NumPy array)
# Create a 100x100 pixel grayscale image with random pixel values
fake_image = np.random.rand(100, 100)

print("Original Image Array Shape:", fake_image.shape)
print("Original Image Data Type:", fake_image.dtype)
print("\nSample of original image pixel values:")
print(fake_image[:5, :5]) # Print a 5x5 sub-section of the array
```

# step:2 Visualize the Original Image

```python
# Create a figure to display our images
plt.figure(figsize=(12, 6))

# Display the original fake image
plt.subplot(1, 3, 1)
plt.imshow(fake_image, cmap='gray')
plt.title('1. Original Image')
plt.axis('off') # Hide the axes for a cleaner look
```

# step:3 Perform Transformations

```python
# 2. Perform Transformations
# Invert the image by subtracting from the maximum value (1.0)
inverted_image = 1.0 - fake_image

# Crop the image by selecting a central portion using array slicing
cropped_image = fake_image[25:75, 25:75]

# Print the shapes to show the change in size
print("Inverted Image Array Shape:", inverted_image.shape)
print("Cropped Image Array Shape:", cropped_image.shape)
```

# step 4. Visualize the Transformed Images

```python
# Display the inverted image
plt.subplot(1, 3, 2)
plt.imshow(inverted_image, cmap='gray')
plt.title('2. Inverted Image')
plt.axis('off')

# Display the cropped image
plt.subplot(1, 3, 3)
plt.imshow(cropped_image, cmap='gray')
plt.title('3. Cropped Image')
plt.axis('off')

plt.tight_layout() # Adjust subplots to give a nice layout
plt.show()
```
