# Visualizing a Derivative

 # Plot a function and its derivative to get an intuitive feel for the concept

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the function
def f(x):
    return x**2   # you can change this to np.sin(x), np.exp(x), etc.

# Step 2: Compute derivative numerically
x = np.linspace(-10, 10, 400)
y = f(x)

# Numerical derivative using numpy's gradient function
dy_dx = np.gradient(y, x)

# Step 3: Plot the function and its derivative
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = x', color='blue')
plt.plot(x, dy_dx, label="f'(x) = 2x (Derivative)", color='red', linestyle='--')

# Step 4: Add labels and legend
plt.title('Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
