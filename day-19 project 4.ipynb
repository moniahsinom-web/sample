 # Visualizing a 3D Gradient

# Plot a 3D surface for a function like f(x,y)=x2+y2 and a vector showing its gradient at a specific point

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Define the function and its gradient
def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    df_dx = 2*x
    df_dy = 2*y
    return df_dx, df_dy

# Step 2: Create meshgrid for surface plot
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Step 3: Choose a specific point for gradient
x0, y0 = 1, 1   # point of interest
z0 = f(x0, y0)
gx, gy = grad_f(x0, y0)

# Step 4: Plot the surface
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter(x0, y0, z0, color='red', s=50, label='Point (1,1)')
ax.quiver(x0, y0, z0, gx, gy, 0, color='red', length=1, arrow_length_ratio=0.2, linewidth=2)

# Step 5: Labels and title
ax.set_title("3D Surface of f(x, y) = x + y with Gradient Vector", pad=20)
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")   #  corrected line
ax.set_zlabel("f(x, y)")
ax.legend()
plt.show()

```
