# Finding the Minimum of a Function Using Gradient Descent

# Use a simple gradient descent implementation to find the minimum of a single-variable function like f(x)=x4−3x3+2

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the function and its derivative
def f(x):
    return x**4 - 3*x**3 + 2

def df(x):
    return 4*x**3 - 9*x**2

# Step 2: Initialize parameters
x = 0.0           # starting point
alpha = 0.01      # learning rate
epochs = 1000     # number of iterations

x_history = []    # to store progress
y_history = []

# Step 3: Gradient Descent loop
for i in range(epochs):
    grad = df(x)
    x = x - alpha * grad
    x_history.append(x)
    y_history.append(f(x))

# Step 4: Visualization
x_vals = np.linspace(-1, 4, 200)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="f(x) = x⁴ - 3x³ + 2", color='blue')
plt.scatter(x_history[-1], f(x_history[-1]), color='red', label='Minimum found')
plt.title("Finding Minimum using Gradient Descent")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(f"Minimum found at x = {x_history[-1]:.4f}")
print(f"f(x) = {f(x_history[-1]):.4f}")
```
