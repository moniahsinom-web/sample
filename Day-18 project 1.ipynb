```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Define a Grid of Vectors

```python
x = np.linspace(-2, 2, 15)
y = np.linspace(-2, 2, 15)
X, Y = np.meshgrid(x, y)

# Stack into 2D vectors (each column is a vector)
vectors = np.vstack([X.ravel(), Y.ravel()])
```

# 2. Define a Transformation Matrix (rotation + scaling)

```python
theta = np.radians(30)  # rotate by 30 degrees
scale = 1.2             # scaling factor
transformation_matrix = scale * np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])
```

```python
# 3. Perform Transformation
transformed_vectors = transformation_matrix @ vectors
```

# 4. Visualize Original and Transformed Grids

```python
plt.figure(figsize=(8, 8))
plt.title("Linear Transformation Visualization", fontsize=14)

# Plot original grid
plt.quiver(vectors[0], vectors[1], np.zeros_like(vectors[0]), np.zeros_like(vectors[1]),
           color='gray', alpha=0.3, scale=1, scale_units='xy')

# Plot transformed vectors as arrows from origin
plt.quiver(np.zeros_like(vectors[0]), np.zeros_like(vectors[1]),
           vectors[0], vectors[1], color='blue', alpha=0.5, label='Original Grid')

plt.quiver(np.zeros_like(vectors[0]), np.zeros_like(vectors[1]),
           transformed_vectors[0], transformed_vectors[1], color='red', alpha=0.6, label='Transformed Grid')
```

```python
# 5. Find Eigenvectors and Eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(transformation_matrix)
```

# 6. Visualize Eigenvectors

```python
origin = np.array([[0, 0], [0, 0]])  # origin point
plt.quiver(*origin, eigenvectors[0, :], eigenvectors[1, :],
           color='green', angles='xy', scale_units='xy', scale=1, width=0.005, label='Eigenvectors')
```

```python
# Beautify plot
plt.legend()
plt.axis('equal')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```
